## Чебышёвская vs. равномерная сетка

```python
deg = 20
n_points = 100
x = np.linspace(-1, 1, num=n_points, endpoint=True)
f = 1. / (1 + 25 * x**2)

x_test = np.linspace(-1, 1, 1000, endpoint=True)
f_test = 1. / (1 + 25 * x_test**2)

w = np.polyfit(x, f, deg)
f_est_uniform = np.polyval(w, x_test)

x_cheb = np.cos(np.pi * np.arange(1, n_points+1) / n_points)
f_cheb = 1. / (1 + 25 * x_cheb**2)
w_cheb = np.polyfit(x_cheb, f_cheb, deg)
f_est_cheb = np.polyval(w_cheb, x_test)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x_test, f_est_cheb, label="Cheb")
plt.plot(x_test, f_est_uniform, label="Uniform")
plt.legend()
```

```python
degs = np.array([i for i in range(1, 100)])
error_uniform = np.zeros(degs.shape[0])

for i, d in enumerate(degs):
    w_polyfit = np.polyfit(x, f, deg=d)
    f_polyval = np.polyval(w_polyfit, x_test)
    error_uniform[i] = np.linalg.norm(f_polyval - f_test)
```

```python
plt.semilogy(error_uniform)
plt.xlabel("Degree of polynomial")
_ = plt.ylabel("Approximation error")
```

## Может ли процесс Арнольди нам помочь?

```python
from numba import jit

@jit(nopython=True)
def arnoldi_polyfit(x, f, deg):
    n = x.shape[0]
    Q = np.zeros((n, deg+1));
    Q[:, 0] = 1. / np.sqrt(n)
    H = np.zeros((deg+1, deg))
    for d in range(deg):
        q = x * Q[:, d]
        for k in range(d+1):
            H[k, d] = Q[:, k] @ q
            q = q - H[k, d] * Q[:, k]
        H[d+1, d] = np.linalg.norm(q)
        Q[:, d+1] = q / H[d+1, d]
    # w = np.linalg.lstsq(Q, f, rcond=None)[0]
    w = Q.T @ f
    return w, H

@jit(nopython=True)
def arnoldi_polyval(x_test, w, H):
    M = x_test.shape[0]
    n = H.shape[1]
    Ht = np.zeros((M, n+1))
    Ht[:, 0] = 1 / np.sqrt(M)
    for k in range(n):
        ht = x_test * Ht[:, k]
        for j in range(k+1):
            ht = ht - H[j, k] * Ht[:, j]
        Ht[:, k+1] = ht / H[k+1, k]
#     print(np.linalg.norm(Ht.T @ Ht - np.eye(n+1)))
    return Ht @ w
```

```python
error_arnoldi_cheb = np.zeros((degs.shape[0],))

for i, d in enumerate(degs):
    w, H = arnoldi_polyfit(x_cheb, f_cheb, d)
    f_arnoldi_polyval_cheb = arnoldi_polyval(x_test, w, H)
    error_arnoldi_cheb[i] = np.linalg.norm(f_arnoldi_polyval_cheb - f_test)
```

```python
plt.semilogy(error_uniform, label="Uniform")
plt.semilogy(error_cheb, label="Cheb")
plt.semilogy(error_arnoldi_cheb, label="Cheb Arnoldi")
plt.xlabel("Degree of polynomial")
_ = plt.ylabel("Approximation error")
_ = plt.legend()
```

## Предобуславливатели

```python
n = 100
A = np.random.randn(n, n)
Q, _ = np.linalg.qr(A)
# D = np.diag(np.array([1e-4, 1] + [*np.arange(95, 100, 0.04)]))
# D = np.diag(np.linspace(1e-4, 100, num=n))
D = np.diag(np.linspace(1, 1000, num=n))
print(D.shape)
A = Q @ D @ Q.T

rhs = np.random.randn(n)

```

```python
import scipy.sparse.linalg as spsplin

cg_cl = []
def callback(x):
    cg_cl.append(x.copy())
x = spsplin.cg(A, rhs, callback=callback)[0]
# %timeit spsplin.cg(A, rhs, callback=callback)[0]
print(np.linalg.norm(A @ x - rhs))
```

```python
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl])
```

```python
eigvals = np.linalg.eigvalsh(A)
plt.scatter(eigvals, np.ones_like(eigvals))
plt.xscale("log")
print(max(eigvals) / min(eigvals))
```

```python
# Test Jacobi
jac_prec = lambda x: x / np.diag(A)
cg_cl_jac_prec = []
def callback(x):
    cg_cl_jac_prec.append(x.copy())
x = spsplin.cg(A, rhs, callback=callback, M=spsplin.LinearOperator((n, n), matvec=jac_prec))[0]
print(np.linalg.norm(A @ x - rhs))
```

```python
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl], label="CG")
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl_jac_prec], label="Jac CG")
plt.legend()
```

```python
# Test Gauss-Seidel
gs_prec = lambda x: np.linalg.solve(np.tril(A), x)
cg_cl_gs_prec = []
def callback(x):
    cg_cl_gs_prec.append(x.copy())
x = spsplin.cg(A, rhs, callback=callback, M=spsplin.LinearOperator((n, n), matvec=gs_prec))[0]
print(np.linalg.norm(A @ x - rhs))

plt.plot([np.linalg.norm(A @ x - rhs) for x in cg_cl_gs_prec])
```

```python
# ILU

invA_approx = spsplin.spilu(A, drop_tol=1e-3)
ilu_prec = lambda x: invA_approx.solve(x)
cg_cl_ilu_prec = []
def callback(x):
    cg_cl_ilu_prec.append(x.copy())
x = spsplin.cg(A, rhs, callback=callback, M=spsplin.LinearOperator((n, n), matvec=ilu_prec))[0]
# %timeit spsplin.cg(A, rhs, callback=callback, M=spsplin.LinearOperator((n, n), matvec=ilu_prec))[0]
print(np.linalg.norm(A @ x - rhs))
```

```python
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl], label="CG")
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl_jac_prec], label="Jac CG")
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl_ilu_prec], label="ILU CG")
plt.legend()
```

```python
eigvals = np.linalg.eigvalsh(np.array(invA_approx.solve(A)))
print(eigvals)
plt.scatter(eigvals, np.ones_like(eigvals))
print(max(eigvals) / min(eigvals))
```

```python
import scipy.sparse as spsp
n = 60

ex = np.ones(n);
# A = -sp.sparse.spdiags(np.vstack((ex,  -(2 + 1./n)*ex, (1 + 1./n) * ex)), [-1, 0, 1], n, n, 'csr') 
rhs = np.random.randn(n*n)

ee = spsp.eye(n)
A = -spsp.spdiags(np.vstack((ex,  -(2 + 1./n)*ex, 1. * ex)), [-1, 0, 1], n, n, 'csr')
A = spsp.kron(A, ee) + spsp.kron(ee, A)
plt.spy(A, markersize=1)
print(A.shape)
```

```python
cg_cl = []
def callback(x):
    cg_cl.append(x.copy())
x = spsplin.cg(A, rhs, callback=callback, tol=1e-8)[0]
print(np.linalg.norm(A @ x - rhs))
```

```python
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl])
```

```python
invA_approx = spsplin.spilu(A, drop_tol=1e-2)
ilu_prec = lambda x: invA_approx.solve(x)
cg_cl_ilu_prec = []
def callback(x):
    cg_cl_ilu_prec.append(x.copy())
x = spsplin.cg(A, rhs, tol=1e-8, callback=callback, M=spsplin.LinearOperator((n**2, n**2), matvec=ilu_prec))[0]
print(np.linalg.norm(A @ x - rhs))
```

```python
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl], label="CG")
plt.semilogy([np.linalg.norm(A @ x - rhs) for x in cg_cl_ilu_prec], label="ILU CG")
plt.legend()
```
